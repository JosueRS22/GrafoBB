```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()

personajes = [
    "Walter White", "Jesse Pinkman", "Skyler White", "Hank Schrader", "Marie Schrader",
    "Saul Goodman", "Gustavo Fring", "Mike Ehrmantraut", "Tuco Salamanca", "Hector Salamanca",
    "Jane Margolis", "Ted Beneke", "Lydia Rodarte-Quayle", "Todd Alquist", "Jack Welker",
    "Andrea Cantillo", "Brock Cantillo", "Gale Boetticher", "Victor", "Tyrus Kitt",
    "Walter White Jr."
]

G.add_nodes_from(personajes)
```

```python
relaciones = [
    ("Walter White", "Jesse Pinkman", "Socios"),
    ("Walter White", "Skyler White", "Esposo"),
    ("Walter White", "Hank Schrader", "Cuñado"),
    ("Walter White", "Marie Schrader", "Cuñado"),
    ("Walter White", "Saul Goodman", "Cliente"),
    ("Walter White", "Gustavo Fring", "Socio"),
    ("Walter White", "Mike Ehrmantraut", "Empleado"),
    ("Walter White", "Lydia Rodarte-Quayle", "Colaboradora"),
    ("Walter White", "Todd Alquist", "Socio"),
    ("Walter White", "Gale Boetticher", "Colaborador"),
    ("Walter White", "Victor", "Asociado"),
    ("Walter White", "Tyrus Kitt", "Asociado"),
    ("Walter White", "Walter White Jr.", "Padre"),
    ("Jesse Pinkman", "Jane Margolis", "Novio"),
    ("Jesse Pinkman", "Saul Goodman", "Cliente"),
    ("Jesse Pinkman", "Andrea Cantillo", "Novio"),
    ("Jesse Pinkman", "Brock Cantillo", "Figura Paterna"),
    ("Jesse Pinkman", "Mike Ehrmantraut", "Socio"),
    ("Jesse Pinkman", "Gustavo Fring", "Empleado"),
    ("Jesse Pinkman", "Todd Alquist", "Colega"),
    ("Skyler White", "Hank Schrader", "Hermana"),
    ("Skyler White", "Marie Schrader", "Hermana"),
    ("Skyler White", "Ted Beneke", "Amante"),
    ("Skyler White", "Walter White Jr.", "Madre"),
    ("Hank Schrader", "Marie Schrader", "Esposo"),
    ("Saul Goodman", "Mike Ehrmantraut", "Colaborador"),
    ("Gustavo Fring", "Mike Ehrmantraut", "Jefe"),
    ("Gustavo Fring", "Gale Boetticher", "Empleador"),
    ("Gustavo Fring", "Victor", "Empleador"),
    ("Gustavo Fring", "Tyrus Kitt", "Empleador"),
    ("Mike Ehrmantraut", "Gustavo Fring", "Empleado"),
    ("Mike Ehrmantraut", "Saul Goodman", "Colaborador"),
    ("Tuco Salamanca", "Hector Salamanca", "Sobrino"),
    ("Tuco Salamanca", "Walter White", "Enemigo"),
    ("Tuco Salamanca", "Jesse Pinkman", "Enemigo"),
    ("Hector Salamanca", "Walter White", "Enemigo"),
    ("Hector Salamanca", "Gustavo Fring", "Enemigo"),
    ("Todd Alquist", "Lydia Rodarte-Quayle", "Colaborador"),
    ("Todd Alquist", "Jack Welker", "Sobrino"),
    ("Jack Welker", "Walter White", "Enemigo"),
    ("Jack Welker", "Jesse Pinkman", "Prisionero"),
    ("Andrea Cantillo", "Brock Cantillo", "Madre")
]

for u, v, relacion in relaciones:
    G.add_edge(u, v, label=relacion)
    
excentricidades = nx.eccentricity(G)

radio = nx.radius(G)
diametro = nx.diameter(G)
centro = nx.center(G)

print("Excentricidades de los nodos:")
for nodo, excentricidad in excentricidades.items():
    print(f'Nodo {nodo}: Excentricidad {excentricidad}')

print(f"\nRadio del grafo: {radio}")
print(f"Diámetro del grafo: {diametro}")
print(f"Centro del grafo: {centro}")
```

```python
pos = nx.spring_layout(G)
plt.figure(figsize=(15, 15))
nx.draw(G, pos, with_labels=True, node_size=3000, node_color="skyblue", font_size=10, font_weight="bold", edge_color="gray")

edge_labels = nx.get_edge_attributes(G, 'label')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red')

plt.title("Grafo de Breaking Bad")
plt.show()
```
